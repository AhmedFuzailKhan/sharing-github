```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('shopping_data.csv')
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
Genre_LE = le.fit_transform(df.Genre)
```

```python
df.Genre = Genre_LE
```

```python
df
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import pairwise_distances
```

```python
model  = AgglomerativeClustering(n_clusters=4, linkage="average", affinity="cosine")
```

```python
model.fit(df.drop(['CustomerID'], axis=1))
```

```python
result = pd.Series(data=model.labels_ , name='result')
```

```python
result.value_counts()
```

```python
pred1 = model.fit_predict(df)
```
