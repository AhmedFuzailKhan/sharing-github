```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
animals = pd.read_csv('horse.csv')
```

```python
animals.head()
```

```python
target = animals['outcome']
```

```python
target.unique()
```

```python
animals = animals.drop(['outcome'], axis=1)
```

```python
category_variables = ['surgery','age','temp_of_extremities', 'peripheral_pulse', 'mucous_membrane', 'capillary_refill_time','pain', 'peristalsis', 'abdominal_distention', 
                     "nasogastric_tube" , "nasogastric_reflux", "rectal_exam_feces", "abdomen", "abdomo_appearance", "surgical_lesion", "cp_data" ]
```

```python
animals.head()
```

```python
animals[category_variables]
```

```python
for category in category_variables: 
    #animals[category] = pd.get_dummies(animals[category])
    animals = pd.concat([animals,pd.get_dummies(animals[category], prefix=category , drop_first=True )],axis=1)
    animals.drop([category], inplace = True , axis=1 )
```

```python
animals.info()
```

```python
animals.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
target = le.fit_transform(target)
```

```python
X_train,X_test,y_train,y_test = train_test_split(animals, target, random_state=42, test_size=0.2)
```

```python
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imp_most_freq = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
```

```python
X_train = imp_most_freq.fit_transform(X_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DT_Classif = DecisionTreeClassifier(random_state=42)
```
